# Data Pre-processing

```python
#checking the missing values
print("The % of Missing Value")
for col in raindata.columns:
    print(col+" : "+str(round(raindata[col].isnull().sum()/raindata.shape[0]*100,2))+"%")
```

```python
#Dropping Columns that Contain Missing Value near close to 30%
raindata.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm'],axis=1,inplace=True)
raindata.info()
```

```python
#numerical variables name
num=[col 
     for col in raindata.columns
        if raindata[col].dtype!='O']
num=num[1:]
print(num)
```

```python
#categorical variables name
cat=[col
    for col in raindata.columns
        if raindata[col].dtype=="O"]
print(cat)
```

```python
#Imputing missing value in numerical variabel with its median
for value in [raindata]:
    for col in num:
        col_median=raindata[col].median()
        value[col].fillna(col_median, inplace=True)

raindata[num].isnull().sum()
```

```python
#Imputing missing value in categorical variable with its mode
for value in [raindata]:
    for col in cat:
        col_mode=raindata[col].mode()[0]
        value[col].fillna(col_mode, inplace=True)
        
raindata[cat].isnull().sum()
```

# Hence we imputed missing values in both types of variables.

# Label Encoding for Target Variable

```python
Y=raindata.iloc[:,16]
Y
```

```python
Y.replace({'No':0,'Yes':1},inplace=True)
Y
```

# One-Hot Encoding for Categorical Predictor Variables

```python
X=raindata.drop(['RainTomorrow'],axis=1)
X
```

```python
X=pd.get_dummies(X)
X
```

# Scaling The Data Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=pd.DataFrame(scaler.transform(X), 
                      index=X.index,
                     columns=X.columns)
```

```python
X_scaled
```

# Data Modeling

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_scaled,Y,test_size=0.2,random_state=50)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
lr.score(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypred_lr=lr.predict(xtest)
accuracy_score(ytest,ypred_lr)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
lr_matrix=confusion_matrix(ytest,ypred_lr)
print(lr_matrix)
```

```python
sns.heatmap(lr_matrix,annot=True,fmt='d')
plt.title("Confusion Matrix Using Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred_lr))
```
